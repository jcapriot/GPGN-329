## Question 1

In class we said that the plane wave was a valid solution to Maxwell's equations in conductive medium:

That is the wave equation for the electric field,
$$\nabla^2 \vec{E} = \mu \sigma \frac{\partial \vec{E}}{\partial t}+ \mu\varepsilon \frac{\partial^2 \vec{E}}{\partial t^2}$$
admitted plane wave solutions:
$$\vec{E}(z, r) = \vec{\tilde{E}}_0 e^{i(\omega t - \tilde{k} z)}$$
such that
$$\tilde{k}^2 = \mu\varepsilon\omega^2 - i\mu\sigma\omega$$.

This is similar to the free-space solutions, except that now $\tilde{k}$ and $\vec{\tilde{E}}_0$ were (in general) complex valued quantities.

We then defined $\tilde{k}$ in terms of it's real and NEGATIVE imaginary parts.
$$\tilde{k} = k_r - i k_i$$

### Part a)

We said that the values for $k_r$ and $k_i$ were:

\begin{align}
k_r &= \omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg]^{1/2}\\
k_i &= \omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg]^{1/2}
\end{align}

Verify that with these definitions that:
$$\tilde{k}^2 = \mu\varepsilon\omega^2 - i\mu\sigma\omega$$

#### Solution

Can do this algebraically:

\begin{align}
(k_r - i k_i)^2 &= \mu\varepsilon\omega^2 - i\mu\sigma\omega\\
k_r^2 - k_i^2 - 2ik_rk_i &= \mu\varepsilon\omega^2 - i\mu\sigma\omega
\end{align}

which gives two equations, one for the real, and one for the imaginary component:

\begin{align}
k_r^2 - k_i^2 &= \mu\varepsilon\omega^2\\
2k_rk_i &= \mu\sigma\omega
\end{align}

So we need to verify both of them, the first is:

\begin{align}
k_r^2 - k_i^2 &= \mu\varepsilon\omega^2\\
\bigg(\omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg]^{1/2}\bigg)^2 - \bigg(\omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg]^{1/2}\bigg)^2 &=\\
\omega^2 \frac{\varepsilon \mu}{2}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg] - \omega^2\frac{\varepsilon \mu}{2}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg] &=\\
\omega^2 \frac{\varepsilon \mu}{2}\bigg[\bigg(\cancel{\sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2}} + 1\bigg) - \bigg(\cancel{\sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2}} - 1\bigg)\bigg] &=\\
\omega^2 \frac{\varepsilon \mu}{2}\bigg[2\bigg] &=\\
\omega^2 \varepsilon \mu &= \mu\varepsilon\omega^2
\end{align}

Which checks out, the second is:

\begin{align}
2k_rk_i &= \mu\sigma\omega\\
2 \bigg(\omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg]^{1/2}\bigg)\bigg(\omega \sqrt{\frac{\varepsilon \mu}{2}}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg]^{1/2}\bigg) &=\\
\omega^2 \varepsilon \mu\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg]^{1/2}\bigg[ \sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg]^{1/2} &=\\
\omega^2 \varepsilon \mu\bigg[\bigg(\sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} + 1\bigg)\bigg(\sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2} - 1\bigg)\bigg]^{1/2} &=\\
\omega^2 \varepsilon \mu\bigg[\bigg(\sqrt{1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2}\bigg)^2 - 1^2\bigg]^{1/2} &=\\
\omega^2 \varepsilon \mu\bigg[1 + \big(\frac{\sigma}{\varepsilon\omega}\big)^2 - 1\bigg]^{1/2} &=\\
\omega^2 \varepsilon \mu\bigg[\big(\frac{\sigma}{\varepsilon\omega}\big)^2\bigg]^{1/2} &=\\
\omega^2 \varepsilon \mu\frac{\sigma}{\varepsilon\omega} &=\\
\omega \mu \sigma &= \mu\sigma\omega
\end{align}

which also checks out

### Part b

We said that there were two regimes we often work at in Geophysical applications, the wave regime where $\sigma << \omega \varepsilon$ and the quasi-static regime $\sigma >> \omega \varepsilon$.

In the wave regime $\sigma << \omega \varepsilon$ these general equations for the complex wavenumber reduce to:
$$k_{r,wave} = \omega \sqrt{\mu \varepsilon}$$
$$k_{i,wave} = \frac{\sigma}{2}\sqrt{\frac{\mu}{\varepsilon}}$$

In the quasitatic regime ($\sigma >> \omega \varepsilon$), $k_r$ and $k_i$ reduce to:

$$k_r\approx k_i \approx k_{quasi} = \sqrt{\frac{\mu \sigma \omega}{2}}$$

You probably are wondering, how much bigger or smaller does conductivity have to be than the product of permitivity and frequency.

Now let's give a little quantification about when each of those regimes apply.

#### ii)

#### Solution

```python
def k_r_func(sigma, mu, eps, omega):
    k_r = (omega
           * np.sqrt(eps * mu/ 2)
           * np.sqrt(np.sqrt(1 + (sigma/(eps * omega))**2) + 1)
          )
    return k_r

def k_i_func(sigma, mu, eps, omega):
    k_i = (omega
           * np.sqrt(eps * mu/ 2)
           * np.sqrt(np.sqrt(1 + (sigma/(eps * omega))**2) - 1)
          )
    return k_i

def k_quasi_func(sigma, mu, eps, omega):
    return np.sqrt(mu * sigma * omega / 2)

def k_r_wave_func(sigma, mu, eps, omega):
    return omega * np.sqrt(mu * eps)

def k_i_wave_func(sigma, mu, eps, omega):
    return sigma/2 * np.sqrt(mu/eps) * np.ones_like(omega)
```

```python
sigma = 0.01

ratio = np.logspace(-3, 3)
omega = sigma/(epsilon_0 * ratio)
k_r = k_r_func(sigma, mu_0, epsilon_0, omega)
k_i = k_i_func(sigma, mu_0, epsilon_0, omega)

k_quasi = k_quasi_func(sigma, mu_0, epsilon_0, omega)
k_r_wave = k_r_wave_func(sigma, mu_0, epsilon_0, omega)
k_i_wave = k_i_wave_func(sigma, mu_0, epsilon_0, omega)
```

```python
plt.loglog(ratio, k_r, label=r'$k_r$')
plt.loglog(ratio, k_i, label=r'$k_i$')
plt.loglog(ratio, k_r_wave, 'C0--', label=r'$k_{r,wave}$')
plt.loglog(ratio, k_i_wave, 'C1--', label=r'$k_{i,wave}$')

plt.legend()
plt.xlabel(r'$\frac{\sigma}{\epsilon\omega}$')
plt.ylabel(r'Wavenumber $m^{-1}$')
```

#### b)

There's many different ways people could have solved this one, so long as they get plots that "look" correct it's okay. Also you can be pretty forgiving on this one.

```python
def k_func(sigma, mu, eps, omega):
    k = (
        k_r_func(sigma, mu, eps, omega)
        - 1j * k_i_func(sigma, mu, eps, omega)
    )
    return k

def a_func(sigma, mu, eps, omega):
    return omega * mu / k_func(sigma, mu, eps, omega)
```

```python
def impedance(sigmas, mus, epsilons, thicknesses, omega):
    sigma = sigmas[0]
    mu = mus[0]
    eps = epsilons[0]
    alpha = a_func(sigma, mu, eps, omega)
    if len(sigmas) == 1:
        return -alpha

    # get the value at the top of the next layer
    Z = impedance(sigmas[1:], mus[1:], epsilons[1:], thicknesses[1:], omega)
    
    h = thicknesses[0]
    k = k_func(sigma, mu, eps, omega)
    tanh = np.tanh(1j * k * h)

    ratio = Z/alpha
    
    top = ratio - tanh
    bot = 1 - ratio*tanh

    return alpha * top/bot
```

```python
sigmas = [0.2, 0.5, 0.1]
mus = [mu_0, mu_0, mu_0]
eps = [epsilon_0, epsilon_0, epsilon_0]
h = [100, 200]

f = np.logspace(-4, 4)

omegas = 2 * np.pi * f

Z0 = impedance(sigmas, mus, eps, h, omegas)
```

```python
plt.loglog(f, -Z0.real, label='real')
plt.loglog(f, -Z0.imag, label='imag')
plt.legend()
```

\begin{align}
\tilde{Z} &= -\alpha = \frac{-\mu\omega}{\tilde{k}}\\

 &= \frac{\mu\omega}{-ke^{i\theta_k}}\\

&= \frac{\mu\omega}{k}e^{-i(\theta_k+\pi)}\\
angle(Z e^{i\theta_z})&= angle(\frac{\mu\omega}{k}e^{-i(\theta_k+\pi)})\\

&\approx-(-\pi/4+\pi)\\
