### Coulomb's Law

EM phenomena are produced and detected by the displacement of a fundamental constituent of matter called electric charge.

Coulomb's law describes the force acting upon one charge due to another charge.

$$\vec{F} = k_e \frac{q_1 q_2}{r^2}\hat{r} $$

* Coulomb's constant $k_e = \frac{1}{4\pi\varepsilon_0}= 8.99 \times 10^9 Nm^2c^{-2}$
* $q_1$, $q_2$ the size of each charge (can be positive or negative)
* $r$ distance between $q_1$ and $q_2$
* $\hat{r}$ points from the source charge to the target charge. I.E. if we wanted to know the force on $q_1$ due to $q_2$, $\hat{r}$ would point from charge 2 to charge 1.
* $\vec{F}$ force of interaction.
* $F_{12} = -F_{21}$ the forces acting on the charges are equal and opposite each other.

### Electrical field:
The source charge generates an vector field which pervades all of space. The detector charge reacts to the local value of this field by experience a force

$$\vec{E} = k_e \frac{q}{r^2}\hat{r}$$
$$\vec{F} = q_{detector} \vec{E}$$

* units of Volts/m = Newtons/Coulomb = Tesla m /s

A bit of evidence that EM waves exist:

* Electrical fields generated by electrical charges
* Electric fields cause charges to move
* Magnetic fields generated by moving electrical charges (electric current)

### Electric Flux
What if we measure the amount of field going through a surface (aka the flux of the field)?

$$\partial \phi_E = \vec{E}\cdot \vec{\partial A}$$

This is a vector dot product, equivalent to:
$$\partial \phi_E = |\vec{E}||\vec{\partial A}| \cos\theta$$
where $\theta$ is the angle between the vectors $\vec{E}$ and $\vec{n}$

* $\theta=0$ $\rightarrow$ field is perpendicular to the surface $\rightarrow$ Maximum Flux!
* $\theta=\frac{\pi}{2}$ $\rightarrow$ field is parallel to the surface $\rightarrow$ No flux!

Q: What is the electric flux due to a point charge?

Take a spherical surface at a radius $r$ from the point charge, then:

\begin{align}
\partial \phi_E &= \vec{E}\cdot \vec{\partial A}\\
\phi &= \int\vec{E}\cdot\vec{\partial A}
\end{align}
The electric field is constant along the sphere centered at the point charge, and it's direction is perpendicular to the sphere, thus:

\begin{align}
\phi &= E_r(\vec{r}) 4 \pi r^2 \\
\phi &= \frac{q}{4 \pi \varepsilon_0 r^2} 4 \pi r^2\\
\phi &= \frac{q}{\varepsilon_0}
\end{align}
Electric flux due to an electric charge does not depend on its location in space.

If we have multiple electric charges at the center? They sum together:

\begin{align}
\phi_E = \frac{\sum q}{\varepsilon_0}
\end{align}

E.G If there were two charges with a value of $-q$ and $q$, there would be no flux.

So in general we have:

\begin{align}
\phi_E = \oint \vec{E}\cdot\vec{\partial A} = \frac{q_{enclosed}}{\varepsilon_0}
\end{align}

Hey That's Gauss's Law! (which so happens to be Maxwell's first equation)

Example:
Take a Hollow metal sphere with radius $R$ and distribute a charge $Q$ on the sphere. Since it is a metal sphere charges are free to move through it and it will have a uniform density of charges.

What is the electric field at a point outside the sphere? What about a point inside the sphere?

Let's draw a circle at radius $R_A > R$ 
$$\oint \vec{E}\cdot\vec{\partial A} = \frac{q_{enclosed}}{\varepsilon_0} = \frac{Q}{\varepsilon}$$
$$E_r(R_A) 4 \pi R_A^2 = \frac{Q}{\varepsilon_0}$$
$$E = \frac{Q}{4 \pi \varepsilon R_A^2} \hat{r}$$

What is it at a point inside the sphere, $R_B< R$?
$$q_{enc} = 0 \rightarrow \vec{E}(R_B)=0$$

### Magnetic Flux
Now consider a $\vec{B}$ Field aka (Magnetic Flux Density), what is its magnetic flux $\phi_B$?

$$\phi_B = \oint \vec{B}\cdot \vec{\partial A} = \frac{\sum q_{mag}}{k}$$

* $q_{mag}$ would be magnetic "charges", and $k$ is some constant

* What is $\sum q_{mag}$?
    * No magnetic monopoles!!
    * $\sum q_{mag}=0$

Implies:
$$\phi_B = \oint \vec{B}\cdot \vec{\partial A} = 0$$
Which is Maxwell's second equation (a.k.a Gauss's Law for magnetism).

2 down, 2 to go...

Really quick, let's look at these in their differential form, for this we need the divergence theorem:

$$
\int_V \nabla \cdot \vec{F}\partial V = \oint_A \vec{F}\cdot\hat{n}\partial A
$$

Let's look at Gauss's Law (Maxwell's first eq) and substitude some things:

\begin{align}
\phi_E = \oint \vec{E}\cdot\vec{\partial A} &= \frac{q_{enclosed}}{\varepsilon_0}\\
\int_V \nabla \cdot \vec{E} \partial V &= \frac{q_{enclosed}}{\varepsilon_0}\\
\int_V \nabla \cdot \vec{E} \partial V &= \int_V\frac{\rho_e}{\varepsilon_0}\\
\end{align}

Because these integrals are over the same domain, there arguments must also be equal,
which leaves the differential form of Gauss's Law:

$$
\nabla \cdot \vec{E} = \frac{\rho_e}{\varepsilon_0}
$$

Similarly Maxwell's second equation can be written as:

$$
\nabla \cdot \vec{B} = 0
$$

## How does a magnetic field induce current?

Consider a circular coil of wire with a radius $r$, there's a few observations we can make"

1. As we move a bar magnet near this wire, a current is induced, and the direction of the current will change based on if we are moving the magnet toward or away from the coil.
2. We have current only when the magnet is moved, the faster we move the magnet the more induced current
3. If we increase the area of the loop, the induced current also increases for the same movement.

If there is current flowing through the loop, there must be a voltage driving that current. We call this voltage, $emf$ (the electromotive force). Our three observations can be stated as:

\begin{align}
emf &\propto \frac{\partial \vec{B}}{\partial t}\\
&\propto A\\
&\propto \frac{\partial \vec{B} \cdot \vec{A}}{\partial t} = \frac{\partial \phi_B}{\partial t}
\end{align}

So, the electromotive force is proportional to the rate of change of the magnetic flux through the loop.

But what is it's direction? Lenz's law tells us:
"The direction of the electric current induced in a conductor by a changing magnetic field is such that the magnetic field created by the induced current opposes changes in the initial magnetic field."

This statement isn't all too different than Newton's third law. The world resists change, the current induced will be opposite the rate of change of the flux, thus:

$$
emf = -\frac{\partial \phi_B}{\partial t}
$$

Consider the relationship between voltage and the electric field from before, that the voltage was the integral of the electric field along the path:

$$
V = \int_A^B \vec{E}\cdot\vec{\partial l}
$$

Well here then, since electromotive force is a voltage, that would imply that over the path of the wire:

\begin{align}
\oint_l\vec{E}\cdot\vec{\partial l} = emf &= -\frac{\partial\phi_B}{\partial t}\\
&= \oint_S \frac{\partial \vec{B}}{\partial t}\cdot \vec{\partial A}\\
\end{align}


Thus we have Faraday's Law (Maxwell's 3rd Equation):

\begin{align}
\oint_l\vec{E}\cdot\vec{\partial l} = -\oint_S \frac{\partial \vec{B}}{\partial t}\cdot \vec{\partial A}
\end{align}

Let's apply Stoke's Theorem, to obtain this in differential form:
$$
\int_S \nabla \times \vec{F} \cdot \vec{\partial A} = \oint_{dS} \vec{F}\cdot\partial L
$$


Applying to the left hand side of Faraday's Law:

\begin{align}
\oint_l\vec{E}\cdot\vec{\partial l} = \oint_S \nabla\times\vec{E} \cdot \vec{\partial A} = \oint_S -\frac{\partial \vec{B}}{\partial t} \cdot \vec{\partial A}
\end{align}

Again we have integrands over the same domain that are equal to each other, thus there arguments must be equal to each other:

$$
\nabla \times \vec{E} = -\frac{\partial B}{\partial t}
$$

### What else creates magnetic fields?
Consider a current flowing into the screen, the Biot-Savart law relates a *steady state* current to a magnetic field:
$$
\vec{B} = \frac{\mu_o I}{2 \pi r} \hat{\phi}
$$
You can suppose it's direction using the right-hand rule, but essentially it curls around the wire. Put your thumb in the direction of current flow, then the $\vec{B}$ field will curl in the direction of your fingers.

Let's integrate the magnetic field along a loop in the plane of the screen, with a radius of $r$ centered on the wire:
$$
\oint \vec{B}\cdot\vec{\partial l}
$$

The magnetic field is constant along the path, and it's direction matches the direction of our integral, thus:
$$
\oint \vec{B}\cdot\vec{\partial l} = B \oint \partial l = B 2 \pi r
$$
replace $B$ with the Biot-Savart law:
$$
B 2 \pi r = \frac{\mu_o I}{2 \pi r} 2 \pi r
$$
Which then simplifies to:
$$
\oint \vec{B}\cdot\vec{\partial l} = \mu_0 I
$$
Which is Ampere's law.

Let's apply Stoke's theorem to this to get a differential form:

\begin{align}
\oint \vec{B} \cdot\vec{\partial l} &= \oint \nabla \times \vec{B} \cdot \vec{\partial A}\\
&= \mu_0 I = \oint \mu_0 \vec{J} \cdot \vec{\partial A}
\end{align}

Which gives the differential form of Ampere's law:
$$
\nabla \times \vec{B} = \mu_0 \vec{J}
$$

Something shouldn't sit right about this in general though... Take the divergence of Ampere's law, and remember that the divergence of the curl of a vector field is always zero.

\begin{align}
\nabla \cdot (\nabla \times \vec{B}) &= \mu_0 (\nabla \cdot \vec{J})\\
0 &= \nabla\cdot J
\end{align}

This is not true in general! Because we know from the continuity of charge that:
$$\nabla \cdot \vec{J} = -\frac{\partial \rho_e}{\partial t}$$ Something must be wrong. But in all honesty this is not surprising considering we started from the Biot-Savart law which assumed steady state current.

So back to Ampere's law, how can we describe what went wrong with it?
Well let's consider what surface the line integral represents. In general the closed circle integral is the boundary path of any surface, it need not be a flat surface. Ampere's law relates the amount of current going through a surface to the magnetic field. So consider instead of a flat surface between our integration circle, let's replace it with a bag that has a circular opening.

If a current goes through the surface of the bag Ampere's law is fine, but consider instead a wire path with a capacitor. We place our bag such that it only includes one plate of the capacitor. Since current doesn't flow through a capacitor, the amount of current that flows through the surface of the bag is zero. But as the capacitor is charging, current is certainly flowing to the capacitor's plate. What is happening is that charge is accumulating within the bag.

Thus we have that in the case of the bag enclosing a single plate of the capacitor:
$$
\mu_0 I = \mu_0 \frac{\partial Q}{\partial t} = \mu_0 \frac{\varepsilon\partial A}{\varepsilon\partial A}\frac{\partial Q}{\partial t} = \mu_0 \varepsilon \partial A \frac{\partial}{\partial t} \frac{Q}{\varepsilon \partial A}
$$

from Gauss's law:

\begin{align}
\vec{E}\cdot\vec{dA} &= \frac{Q}{\varepsilon}\\
E &= \frac{Q}{\varepsilon \partial A}
\end{align}

Plugging this back in to our bag equation:
$$
\mu_0 \varepsilon \partial A \frac{\partial}{\partial t} \frac{Q}{\varepsilon \partial A} = \mu_0 \varepsilon \partial A \frac{\partial E}{\partial t} =  \mu_0 \varepsilon \frac{\partial (E \partial A}{\partial t} = \mu_o \varepsilon \frac{\partial \phi_E}{\partial t}
$$

So correcting Ampere's law with this term gives:

$$
\oint \vec{B}\cdot\vec{\partial l} = \mu_0 I + \mu_o \varepsilon \frac{\partial \phi_E}{\partial t}
$$
or in terms of the electric field instead of the electric flux:

$$
\oint \vec{B}\cdot\vec{\partial l} = \mu_0 I + \mu_o \varepsilon \iint \frac{\partial \vec{E} }{\partial t}\cdot \vec{\partial A}
$$

Again let's complete this with the differential form, apply stoke's theorem again to the right and we have:

$$
\iint \nabla \times \vec{B} \cdot \vec{\partial S}= \iint \mu_0 \vec{J} \cdot \vec{\partial A} + \mu_o \varepsilon \iint \frac{\partial \vec{E} }{\partial t}\cdot \vec{\partial A}
$$

which implies:
$$
\nabla \times \vec{B} = \mu_0 (\vec{J} + \varepsilon \frac{\partial \vec{E}}{\partial t})
$$

Maxwell's 4th equation: Ampere's law with maxwell's correction.

As an alternative to this derivation, we could've gotten here from the divergence of Ampere's law by simply subtracting off a correction term to make $\nabla \cdot \nabla \vec{B}=0$

That term would be:
$$
\nabla \cdot \nabla \times \vec{B}=\mu_0 \nabla \cdot \vec{J} - \mu_0 \nabla \cdot \vec{J}
$$

replace the second $\nabla \cdot \vec{J}$ with the continuity of charge equation, and insert Gauss's Law:

\begin{align}
\nabla \cdot \nabla  \times \vec{B}&=\mu_0 \nabla \cdot \vec{J} - \mu_0 (-\frac{\partial \rho_e}{\partial t})\\
&=\mu_0 \nabla \cdot \vec{J} + \mu_0 \frac{\partial }{\partial t} \varepsilon \nabla \cdot \vec{E}\\
&=\mu_0 \nabla \cdot (\vec{J} + \varepsilon \frac{\partial \vec{E}}{\partial t})
\end{align}

This correction is often refered to as displacement currents (despite that there is not actually a current).

There should be a bit of satisfying symmetry here. Faraday's law states that changing magnetic fields create electric fields, Ampere's Law with Maxwell's correction states that changing electric fields create magnetic fields.

This symmetry almost begs for there to be a corresponding magnetic "charge" that creates magnetic fields just like the 1st equation says that electric charges create electric fields, but desptie much searching no one has been able to find a magnetic "charge".

Fundamentally all of these phenomenon are created by electric charges, the first equation is self evident for electric fields, but by the 4th equation both moving electric charges and the displacement of electric charges cause magnetic fields.
