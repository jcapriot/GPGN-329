# Hooke's Law

Various materials responds differently to applied stress (remember Mohr's cirle, etc.) The relation between stress and strain is given by a materials constitutive equations. Remember we had constitutive relations in Electromagnetics as well relating $\vec{E}$ to $\vec{J}$ (Ohm's Law) $\vec{H}$ to $\vec{B}$ and $\vec{E}$ to $\vec{D}$!

The simplest materials are linearly elastic materials such that the relation between stress and strain is linear. Which we describe with Hooke's Law.

\begin{align}
\sigma_{ij} = c_{ijkl} \varepsilon_{kl}
\end{align}

$c_{ijkl}$ are the elastic constants of the material (elastic moduli).

Hooke's law is valid for infinitesimal strains, assuming that displacements are small. For example:
* Body waves : 10 $\mu$m displacements, but wavelengths are 10km!

In 1D Hooke's law goes to the simple expression for a spring:
$$F = k x$$
in 1D then, there is 1 unique describing parameter.

In higher dimensions, there are potentially $dim^4$ parameters in the tensor $c_{ijkl}$ ($2^4=16$ in 2D, $3^4=81$ in 3D, however there are a few necessary symmetries we can account for.

First that $\underline\sigma$ and $\underline\varepsilon$ are symmetric, this implies that:
$$c_{ijkl} = c_{jikl} = c_{jilk} = c_{ijlk}$$

Which brings it down to 4 unique parameters in 2D and 36 in 3D.

The second is that strain energy is conserved, implying:
$$c_{ijkl} = c_{klij}$$

Thus there are

| Dimension    | Unique components of $c_{ijkl}$ |
| -------- | ------- |
| 1  | 1    |
| 2 | 3     |
| 3    | 21    |

## Isotropic materials

### Lame's Parameters
If the material is isotropic, then $c_{ijkl}$ has further symmetries, and there are only 2 independent elastic moduli, called $\lambda$ and $\mu$, the Lame parameters.

\begin{align}
c_{ijkl} &= \lambda \delta_{ij}\delta_{kl} + \mu (\delta_{ik}\delta_{jl} + \delta_{il}\delta_{jk})\\
\sigma_{ij} &= \big(\lambda \delta_{ij}\delta_{kl} + \mu (\delta_{ik}\delta_{jl} + \delta_{il}\delta_{jk})\big) \varepsilon_{kl}\\
\sigma_{ij} &= \lambda \delta_{ij}\delta_{kl}\varepsilon_{kl} + \mu (\delta_{ik}\delta_{jl}\varepsilon_{kl} + \delta_{il}\delta_{jk}\varepsilon_{kl}) \\
\sigma_{ij} &= \lambda \varepsilon_{kk}\delta_{ij} + \mu (\delta_{ik}\varepsilon_{jk} + \delta_{il}\varepsilon_{jl})\\
\sigma_{ij} &= \lambda \varepsilon_{kk}\delta_{ij} + \mu (\varepsilon_{ji} + \varepsilon_{ij})\\
\sigma_{ij} &= \lambda \theta\delta_{ij} + 2\mu \epsilon_{ij}\\
\underline{\sigma} &= \lambda trace(\underline\varepsilon) I  + 2\mu \underline\varepsilon\\
\underline{\sigma} &= \lambda \theta I  + 2\mu \underline\varepsilon\\
\end{align}

thus:
\begin{align}
\sigma_{ii} & = \lambda \theta + 2 \mu \varepsilon_{ii} &\\
\sigma_{ij} &= 2 \mu \varepsilon_{ij} &\textrm{ for } i\neq j
\end{align}

### Shear and Bulk Modulus

$\mu$ relates the shear stresses to shear strains, and is also called the **shear modulus**

Let's look at a term $K$, called the **bulk modulus**, first apply a uniform pressure to a body and look at the change in volume of it, the bulk modulus is defined as the negative change in volume due to a change in pressue, per unit volume.

\begin{align}
K = -\frac{\Delta P}{\Delta \theta}
\end{align}

Through Hooke's law in an isotropic medium, we can relate these, and the change in pressue is equal to the negative change in stress.
\begin{align}
\Delta \sigma_{ij} &= -\Delta P \delta_{ij}\\
&= \lambda \Delta\theta \delta_{ij} + 2 \mu \Delta\varepsilon_{ij}\\
-\Delta P \delta{ii} &= \lambda \Delta\theta \delta_{ii} + 2 \mu \Delta\varepsilon_{ii}\\
 &= \lambda \Delta\theta \delta_{ii} + 2 \mu \Delta\theta\\
-3\Delta P &= 3 \lambda \Delta\theta + 2 \mu \Delta\theta\\
-3\Delta P &= 3 \lambda \Delta\theta + 2 \mu \Delta\theta\\
\frac{-\Delta P}{\Delta \theta} &= \frac{3 \lambda  + 2 \mu}{3}\\
K &= \lambda + \frac{2}{3} \mu
\end{align}

So we have a relationship between Lame's first parameter $\lambda$, Lame's second parameter/shear modulus $\mu$ and the bulk modulus $K$.

The Shear modulus describes a material's rigiditiy (how hard is it to bend it?):
$$ \mu \geq 0 $$ meaning that a material can't bend opposite the direction you apply force to it.

The Bulk modulus describes how hard it is to compress a material (How hard is it to squeeze?):
$$ K \geq 0 $$ (a material can't get bigger when you squeeze it).

For example, in an ideal fluid, $\mu=0$ and $K = \lambda$

We can express Hooke's law in terms of the Bulk modulus and Shear modulus:

\begin{align}
\lambda &= K - \frac{2}{3}\mu\\
\sigma_{ij} &= \lambda \theta \delta_{ij} + 2 \mu \varepsilon_{ij}\\
\sigma_{ij} &= (K - \frac{2}{3}\mu) \theta \delta_{ij} + 2 \mu \varepsilon_{ij}\\
\sigma_{ij} &= K \theta \delta_{ij} + 2 \mu (\varepsilon_{ij}-\frac{\theta}{3}\delta_{ij})\\
\end{align}

### Poisson's ratio and Young's Modulus

Poisson's ratio and Young's Modulus describe a material's response to pulling it along one axis.

Consider the relationship between the diagonal components of  along a single axis is:
\begin{align}
\sigma_{ii} &= \lambda (\varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}) + 2 \mu \varepsilon_{ii}\\
\sigma_{11} &= (\lambda + 2 \mu ) \varepsilon_{11} + \lambda \varepsilon_{22} + \lambda \varepsilon_{33}\\
\sigma_{22} &= (\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} + \lambda \varepsilon_{33}\\
\sigma_{33} &= (\lambda + 2 \mu ) \varepsilon_{33} + \lambda \varepsilon_{11} + \lambda \varepsilon_{22}\\
\end{align}

If we apply tenson only along $x_1$, implies $\sigma_{11}\neq0$, $\sigma_{22} = \sigma_{33} = 0$.
\begin{align}
\sigma_{11} &= (\lambda + 2 \mu ) \varepsilon_{11} + \lambda \varepsilon_{22} + \lambda \varepsilon_{33} & (1)\\
0 &= (\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} + \lambda \varepsilon_{33} & (2)\\
0 &= (\lambda + 2 \mu ) \varepsilon_{33} + \lambda \varepsilon_{11} + \lambda \varepsilon_{22} & (3)\\
\end{align}

Which is a system of equations we can solve:
set the last two equal to each other to get relationship between $\varepsilon_{22}$ and $\varepsilon_{33}$
\begin{align}
(\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} + \lambda \varepsilon_{33} &= (\lambda + 2 \mu ) \varepsilon_{33} + \lambda \varepsilon_{11} + \lambda \varepsilon_{22}\\
(\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{33} &= (\lambda + 2 \mu ) \varepsilon_{33} + \lambda \varepsilon_{22}\\
\lambda \varepsilon_{22} + 2 \mu \varepsilon_{22}  + \lambda \varepsilon_{33} &= \lambda\varepsilon_{33} + 2 \mu \varepsilon_{33} + \lambda \varepsilon_{22}\\
2 \mu \varepsilon_{22}  &= 2 \mu \varepsilon_{33}\\
\varepsilon_{22}  &=\varepsilon_{33}
\end{align}

Insert into the second (or third) equation:
\begin{align}
(\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} + \lambda \varepsilon_{33} &= 0\\
(\lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} + \lambda \varepsilon_{22} &= 0\\
(2 \lambda + 2 \mu ) \varepsilon_{22} + \lambda \varepsilon_{11} &= 0\\
2(\lambda + \mu ) \varepsilon_{22} &= -\lambda \varepsilon_{11}\\
\varepsilon_{22} &= \frac{-\lambda}{2 (\lambda + \mu)}  \varepsilon_{11} = \varepsilon_{33}\\
\end{align}

Poisson's ratio is the negative ratio of transverse strain to axial strain, (in this example $\sigma_{22}$ and $\sigma_{33}$ are the transverse strains and $\sigma_{11}$ is the axial strain. As a material is squeezed in one direction, it tends to expand in the directions perpendicular to it, Poisson's ratio is describes this.

$$ \nu = \frac{-\varepsilon_{\perp}}{\varepsilon_{||}}$$

Therefore:
\begin{align}
-\nu \varepsilon_{11} &= \varepsilon_{22} = \varepsilon_{33}\\
-\nu \varepsilon_{11}&= \frac{-\lambda}{2 (\lambda + \mu)}  \varepsilon_{11}\\
\nu &= \frac{\lambda}{2 (\lambda + \mu)}
\end{align}

Young's modulus is the ratio of the tensile stress and axial strain:

$$ E = \frac{\sigma}{\varepsilon}$$

So in our case it is:
\begin{align}
E &= \frac{\sigma_{11}}{\varepsilon_{11}}\\
&= \frac{(\lambda + 2 \mu ) \varepsilon_{11} - \lambda \frac{\lambda}{2 (\lambda + \mu)}  \varepsilon_{11} - \lambda \frac{\lambda}{2 (\lambda + \mu)}  \varepsilon_{11}}{\varepsilon_{11}}\\
&= (\lambda + 2 \mu ) - \frac{\lambda^2}{(\lambda + \mu)}\\
&= \frac{(\lambda + \mu) (\lambda + 2 \mu ) - \lambda^2}{\lambda + \mu}\\
&= \frac{\lambda^2 + 2\mu^2 + 3\mu\lambda - \lambda^2}{\lambda + \mu}\\
&= \mu\frac{3\lambda + 2\mu}{\lambda + \mu}
\end{align}

In a "Poisson solid" $\mu = \lambda$, and
\begin{align}
E&=\mu\frac{3\lambda + 2\mu}{\lambda + \mu}\\
 &=\mu\frac{3\mu + 2\mu}{\mu + \mu}\\
 &=\frac{5\mu}{2} = \frac{5\lambda}{2}
\end{align}
