# Fundamentals of waves

Consider a stretched string, a simple physical system that gives rise to waves.

**_NOTE:_**  This system is physically similar to how seismic waves propogate. The string is a deformable material.

A function that describes the position of the string:
$$
\vec{s}(\vec{x}, t)
$$

The velocity of it's movement (and it's acceleration) are given by simple time derivatives of $\vec{s}$

$$
\vec{v}(\vec{x}, t) = \partial_t \vec{s}(\vec{x}, t)
$$
$$
\vec{a}(\vec{x}, t) = \partial_t^2 \vec{s}(\vec{x}, t)
$$

Consider an idealized version of this string, for instance a guitar string,
that extends infinitely in the x direction.

At $t=0$ the string is straight in response to a tension force $\tau$ exerted along it $(\vec{s}(x, 0)=\vec{0})$. At some later time, $t>0$, it is plucked (displaced away from zero) in the y-direction $(\vec{s}(x, t>0) \parallel \hat{y})$.

Our goal is to be able to describe the y-position of the string as a function of space and time $u(x, t)$.

Focus in on a small segment of the string, say that the left side is at position
$$x_1, y_1$$ and the right side is at a position $$x_2, y_2 = x_1 + \Delta x, y_1 + \Delta y$$.

Note that in this example our function $u$ describes the y component of the string's position as a function of $x$ and $t$, thus we could also say that the left side is at
$$x_1, y_1 = x_1, u(x_1, t)$$ and the right side at $$x_2, y_2 = x_2, u(x_2, t) = x_1 + \Delta x, u(x_1+\Delta x, t)$$ and that our $$\Delta y = u(x_1 + \Delta x,t) - u(x_1, t) = \Delta u$$

For our small segment, let's assume that it's $y$ displacement is much less than it's length, meaning $$\Delta x >> \Delta u.$$ Which implies that it's length is equal to $$L_{segment} \approx \Delta x$$.

Newton's second law:
$$\vec{F} = m\vec{a} $$

Once this small segment is displaced, the tension is still directed along the string, but since the string is no longer solely along the x-axis, the tension force will have both $x$ and $y$ components. Let's look at just the y-component of the string, as that is the direction of it's displacement we are interested in (remember $u$ describes it's y-position).

On the left side:
$$
\tau_{1, y} = -\tau \sin\theta_1
$$
And on the right:
$$
\tau_{2, y} = \tau \sin\theta_2
$$

The net force on the string in the $y$-direction will cause it to accelerate in the $y$-direction according to Newton's second law:

\begin{align}
F_{net, y} = \tau \sin\theta_2 - \tau\sin\theta_1 &= m a_y\\
& = \rho L_{segment} a_y = \rho \Delta x a_y\\
&= \rho \Delta x \frac{\partial_t^2 u(x, t)}{\partial t^2}
\end{align}

If the angles are very small, then we can approximate $\sin(\theta) \approx \tan(\theta) \approx \theta$.

```python
# a bit of graphical evidence here about the above claim.
thetas = np.linspace(-0.001, 0.001, 32)
plt.plot(thetas, np.sin(thetas), marker='*', label=r'$y = \sin\theta$')
plt.plot(thetas, np.tan(thetas), label=r'$y = \tan\theta$', marker='x')
plt.plot(thetas, thetas, label=r'$y = \theta$')
plt.legend()
```

\begin{align}
\tau \sin\theta_2 - \tau\sin\theta_1 &= \rho \Delta x \frac{\partial_t^2 u(x, t)}{\partial t^2}\\
\tau (\tan\theta_2 - \tan\theta_1) &=
\end{align}

At each end, $\tan\theta$ is essentially equal to the slope of the line segment at those points, because $\theta$ was the angle of the string at each end:

$$
\tan\theta_1 = \frac{\partial u(x_1, t)}{\partial x}
$$
$$
\tan\theta_2 = \frac{\partial u(x_2, t)}{\partial x}
$$

then:
\begin{align}
\tau (\tan\theta_2 - \tan\theta_1) &=  \rho \Delta x \frac{\partial_t^2 u(x, t)}{\partial t^2}\\
\tau (\frac{\partial u(x_2, t)}{\partial x} - \frac{\partial u(x_1, t)}{\partial x}) &=\\
\tau (\frac{\partial u(x_1 + \Delta x, t)}{\partial x} - \frac{\partial u(x_1, t)}{\partial x}) &=
\end{align}

Remind yourself of the Taylor series:

$$
f(x + \delta x) = f(x) + f'(x)\delta x + f''(x)\frac{\delta x^2}{2} + ...
$$

for small $\delta x \approx 0$ the higher order terms of $\delta x$ go to zero faster the the lower order terms, and we say:
$$
f(x + \delta x) \approx f(x) + f'(x)\delta x
$$

Let's apply this Taylor series approximation to $f(x) = \partial_x u(x, t)$.

\begin{align}
f(x + \delta x) &\approx f(x) + f'(x)\delta x\\
\partial_x u(x_1 + \Delta x, t) &\approx \partial_x u(x_1, t) + \partial_x^2 u(x_1, t) \Delta x
\end{align}

let's use this to replace the first term of:
\begin{align}
\tau (\partial_x u(x_1 + \Delta x, t) - \partial_x u(x_1, t)) &=  \rho \Delta x \partial_t^2 u(x, t)\\
\tau ((\partial_x u(x_1, t) + \partial_x^2 u(x_1, t) \Delta x) - \partial_x u(x_1, t)) &\approx  \rho \Delta x \partial_t^2 u(x, t)\\
\tau \partial_x^2 u(x_1, t) \Delta x &\approx \rho \Delta x \partial_t^2 u(x, t)\\
\tau \partial_x^2 u(x_1, t)&\approx \rho \Delta x \partial_t^2 u(x, t)\\
\partial_x^2 u(x_1, t)&\approx  \frac{\rho}{\tau} \partial_t^2 u(x, t)
\end{align}


For a sufficiently small $\Delta x$ these to quantities are equal to each other. Let's rewrite this as:

\begin{align}
\frac{\partial^2 u(x, t)}{\partial x^2} &= \frac{1}{v^2}\frac{\partial^2 u(x, t)}{\partial x^2}\\
v &= \sqrt{\frac{\tau}{\rho}}
\end{align}

This is termed a hyperbolic partial differential equation. Second order spatial derivatives are coupled to second order time derivatives. The equation gives rise to waves that propogate with a velocity $v$.

It is called the wave equation!

## Solutions to the wave equation

claim:
Any function of the form $f(x\pm v t)$ is a solution to the wave equation.

Proof:
\begin{align}
\partial_x^2 f(x \pm v t) &= f''(x\pm v t) \\
\partial_t^2 f(x \pm v t) &= (\pm v)^2 f''(x\pm v t) = v^2 f''(x\pm v t)
\end{align}

Plug these into the 1D wave equation:
\begin{align}
\frac{\partial^2 f(x \pm v t)}{\partial x^2} &= \frac{1}{v^2}\frac{\partial^2 f(x \pm v t}{\partial x^2}\\
f''(x\pm v t) &= \frac{1}{v^2} ( v^2 f''(x\pm v t)\\
f''(x\pm v t) &= f''(x\pm v t)\\
\end{align}

A common solution (for good reason) to this equation is a sine wave:
$$f(x\pm v t) = \sin(x \pm v t)
$$

### A general solution
But what really is the general solution to this equation?

Let us look at set two variables, both as a function of $x$ and $t$
$$a = x - vt $$ and $$b = x + vt$$.

What is $\partial_x$? Apply the chain rule.
$$
\frac{\partial}{\partial x} = \frac{\partial a}{\partial x}\frac{\partial}{\partial a} + \frac{\partial b}{\partial x}\frac{\partial}{\partial b}
$$

$$
\frac{\partial}{\partial x} = 1 \frac{\partial}{\partial a} + 1\frac{\partial}{\partial b}
$$

$$
\frac{\partial}{\partial x} =\frac{\partial}{\partial a} + \frac{\partial}{\partial b}
$$

What about $\partial_t$?

$$
\frac{\partial}{\partial t} = \frac{\partial a}{\partial t}\frac{\partial}{\partial a} + \frac{\partial b}{\partial t}\frac{\partial}{\partial b}
$$

$$
\frac{\partial}{\partial t} = -v \frac{\partial}{\partial a} + v\frac{\partial}{\partial b}
$$

$$
\frac{\partial}{\partial t} = v (\frac{\partial}{\partial b} - \frac{\partial}{\partial a})
$$

Lets make use of these operators in the wave equation:
The second spatial derivative of u:
\begin{align}
\partial_x^2 u &= (\partial_a + \partial_b)^2u\\
&=\partial_a^2 u + \partial_b^2u + 2\partial_a\partial_b u
\end{align}
The second time derivative of u:
\begin{align}
\partial_t^2 u &= v^2(\partial_b - \partial_a)^2u\\
&=v^2(\partial_a^2 u + \partial_b^2u - 2\partial_a\partial_b u)
\end{align}

Insert these two into the wave equation:

\begin{align}
\partial_x^2 u &= \frac{1}{v^2} \partial_t^2 u\\
\partial_a^2 u + \partial_b^2u + 2\partial_a\partial_b u &= \frac{1}{v^2} \partial_t^2 u\\
\partial_a^2 u + \partial_b^2u + 2\partial_a\partial_b u &= \frac{1}{v^2} v^2(\partial_a^2 u + \partial_b^2u - 2\partial_a\partial_b u)\\
\partial_a^2 u + \partial_b^2u + 2\partial_a\partial_b u &= \partial_a^2 u + \partial_b^2u - 2\partial_a\partial_b u\\
2\partial_a\partial_b u &= - 2\partial_a\partial_b u\\
4\partial_a\partial_b u &= 0\\
\partial_a\partial_b u &= 0
\end{align}

This partial differential equation has a general solution (called the d'Alembert's solution):

\begin{align}
u(x, t) &= f(a) + g(b)\\
&= f(x - v t) + g(x + vt)
\end{align}

So the solution is in general a right traveling function $f(x - v t)$ and a left traveling function $g(x + vt)$.

## A general sinusoidal solution

consider 
\begin{align}
u(x, t) &= A e^{i (\omega t \pm k x)}\\
&= A \cos(\omega t \pm k x) + i A \sin(\omega t \pm k x)
\end{align}

which is of the form $f(x - vt) + g(x + vt)$, and a valid solution to the wave equation.

How do we describe properties of the wave?

The wave has an amplitude $A$ and a phase $(\omega t \pm k x)$

It has these characteristics summarized below:


| Space || Time ||
 --- | --- | --- | --- |
 |$k$| (angular) wavenumber (rad/m)| $\omega$ | angular frequency (rad/s)|
 |$\xi$| spatial wavenumer (spatial frequency) (1/m)| $f$ | (ordinary) frequency (1/s)|
 |$\lambda$| wavelength (m)| $T$ | period (s)|

 All of these are related to each other, in space:
\begin{align}
k = 2\pi\xi = \frac{2\pi}{\lambda}\\
\xi = \frac{1}{\lambda}
\end{align}

and in time:
\begin{align}
\omega = 2\pi f = \frac{2\pi}{T}\\
f = \frac{1}{T}
\end{align}

what is the velocity of this wave?

$$
u_{xx} = (i k)^2 A e^{i(\omega t \pm kx)}= - k^2 u
$$
$$
u_{tt} = (i \omega)^2 A e^{i(\omega t \pm kx)} = -\omega^2 u
$$

insert into wave equation:
\begin{align}
u_{xx} & = \frac{1}{v^2} u_{tt}\\
- k^2 u &= \frac{1}{v^2} -\omega^2 u\\
k^2 &= \frac{1}{v^2} \omega^2\\
v^2 &= \frac{\omega^2}{k^2} \\
v &= \frac{\omega}{k}
\end{align}

That means:
$$
v = \frac{\omega}{k} = \frac{f}{\xi} = \frac{\lambda}{T}
$$

Q: If velocity is constant, how will wavenumber change as a function of frequency?

$$
k = \frac{\omega}{v}, \hspace{20pt} \omega \uparrow \rightarrow k \uparrow
$$

Often in geophysics we cannot control the velocity of the wave, but we can control its frequency.

# Waves equation in 3D

This wave equation has a natural extension to higher dimensions:

$$
\nabla^2 u = \frac{1}{v^2} \partial_{tt} u
$$
This is the 3D scalar wave equation.

Analogously it's solutions are of the form:
$$
u(\vec{x}, t) = f(\vec{r} \pm v t)
$$
or
$$
u(\vec{x}, t) = f(\omega t \pm \vec{k}\cdot\vec{r})
$$
So now, each spatial dimension has its own wavenumber:
$$\vec{k} = k_x \hat{x} + k_y \hat{y} + k_z\hat{z}$$
and of course $\vec{r}$ is:
$$ \vec{r} = x \hat{x} + y \hat{y} + z \hat{z}$$

Q? what does the gradient of $f$ look like?

$$
\nabla f(\omega t \pm \vec{k}\cdot\vec{r}) = \pm \vec{k}f'(\omega t \pm \vec{k}\cdot \vec{r})
$$
